## Load data

### Subtask:
Load the Mall Customers dataset from the provided CSV file into a pandas DataFrame.


**Reasoning**:
Import the pandas library and load the dataset from the CSV file into a pandas DataFrame.



```python
import pandas as pd

df = pd.read_csv("/content/Mall_Customers.csv")
```

**Reasoning**:
The dataset has been loaded into a pandas DataFrame. Display the first few rows to verify the data has been loaded correctly.



```python
display(df.head())
```

## Preprocess data

### Subtask:
Drop the 'CustomerID' column and encode the 'Gender' column (Male = 0, Female = 1).


**Reasoning**:
Drop the 'CustomerID' column and encode the 'Gender' column as instructed.



```python
df = df.drop('CustomerID', axis=1)
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})
display(df.head())
```

## Feature selection and scaling

### Subtask:
Select the features 'Annual Income (k$)' and 'Spending Score (1-100)', and scale the selected features using StandardScaler.


**Reasoning**:
Select the specified features and scale them using StandardScaler as per the instructions.



```python
from sklearn.preprocessing import StandardScaler

features = df[['Annual Income (k$)', 'Spending Score (1-100)']]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
display(scaled_features[:5])
```

## Determine optimal clusters

### Subtask:
Use the Elbow Method to find the optimal number of clusters, and plot inertia vs. k for k ranging from 1 to 10.


**Reasoning**:
Implement the Elbow Method to find the optimal number of clusters and plot the results as instructed.



```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
for k in range(1, 11):
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    model.fit(scaled_features)
    inertia.append(model.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal k")
plt.show()
```

## Apply kmeans clustering

### Subtask:
Apply the KMeans algorithm with the optimal number of clusters found in the previous step, and assign the cluster labels to the DataFrame.


**Reasoning**:
Apply KMeans clustering with k=5 and assign the cluster labels to the original DataFrame.



```python
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
kmeans.fit(scaled_features)
df['cluster'] = kmeans.predict(scaled_features)
display(df.head())
```

## Visualize clusters

### Subtask:
Create a scatter plot of 'Annual Income (k$)' vs. 'Spending Score (1-100)' and color the data points according to their assigned cluster labels.


**Reasoning**:
Generate a scatter plot to visualize the customer clusters based on 'Annual Income (k$)' and 'Spending Score (1-100)', using different colors for each cluster.



**Reasoning**:
The previous code failed because of a KeyError due to a typo in the column name 'Spending Score (100)'. The correct column name is 'Spending Score (1-100)'. I need to fix the column name in the scatter plot code to correctly access the data.



```python
plt.figure(figsize=(10, 6))
for cluster_id in sorted(df['cluster'].unique()):
    cluster_data = df[df['cluster'] == cluster_id]
    plt.scatter(cluster_data['Annual Income (k$)'], cluster_data['Spending Score (1-100)'], label=f'Cluster {cluster_id}')

plt.title('Customer Segments based on Annual Income and Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Clusters')
plt.grid(True)
plt.show()
```
